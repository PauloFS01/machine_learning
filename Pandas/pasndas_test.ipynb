```python
df_bike = pd.read_csv('../CSVs/SeoulBikeData2.csv')
df_bike.head()
```

```python
#summary functions
df_bike.bike_count.describe() 

#to string datas
df.MainBranch.describe()

#get some particular data summary
df_bike.bike_count.mean()
df_bike.rain.unique()
df_bike.radiation.count()

#maps
bike_means = df_bike.bike_count.mean()
df_bike.bike_count.map(lambda b: b - bike_means)

#apply
def change_count (row):
    row.bike_count = row.bike_count - bike_means
    return row

df_bike.apply(change_count, axis='columns')
```

```python
#simple operators
df.MainBranch + "-" + df.RemoteWork

df_bike.temp + df_bike.humidity
```

```python
#function to call in apply
def rating_radiation(row):
    if row.radiation > 0.5:
        return 2
    elif row.radiation > 1:
        return 0
    elif row.radiation < 0.5:
        return 3
    
stars_rating_radiation = df_bike.apply(rating_radiation, axis='columns')
```

```python
#groupby bike counts for humid days
df_bike.groupby('humidity').bike_count.count().head()

#groupby using apply
df_bike.groupby('humidity').apply(lambda c: c.bike_count.iloc[0])

#df.groupby(['CodingActivities', 'Employment']).apply(lambda c: c.loc[c.YearsCode.idxmax()])
#df.head()

#using agg to pass funcions
humidity_vaules = df_bike.groupby('humidity').radiation.agg([min,max])

#use sorting_values
humidity_vaules.sort_values(by=['min', 'max'], ascending=False)

#agruping by mean
df_bike.groupby('temp').humidity.mean()

#soting by size
df_bike.groupby('temp').size().sort_values(ascending=False)
```
